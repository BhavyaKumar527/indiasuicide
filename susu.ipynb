```python
status = pd.get_dummies(susu_causes['State'], drop_first= True)
susu_causes = pd.concat([susu_causes, status], axis=1)
susu_causes.drop(['State'], axis = 1 , inplace= True)
```

```python
status = pd.get_dummies(susu_causes['Type'], drop_first= True)
susu_causes = pd.concat([susu_causes, status], axis=1)
susu_causes.drop(['Type'], axis = 1 , inplace= True)
```

```python
status = pd.get_dummies(susu_causes['Gender'], drop_first= True)
susu_causes = pd.concat([susu_causes, status], axis=1)
susu_causes.drop(['Gender'], axis = 1 , inplace= True)
```

```python
status = pd.get_dummies(susu_causes['Age_group'], drop_first= True)
susu_causes = pd.concat([susu_causes, status], axis=1)
susu_causes.drop(['Age_group'], axis = 1 , inplace= True)
```

```python
susu_causes.info()
```

```python
X = susu_causes.drop(['Total','Type_code'],axis=1)
y = susu_causes['Total']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.7,test_size=0.3,random_state=100,shuffle=True)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
rfe = RFE(model, n_features_to_select=30)
rfe.fit(X_train,y_train)
```

```python
col = X_train.columns[rfe.support_]
col
```

```python
X_train_rfe = X_train[col]
X_train_rfe
```

```python
import statsmodels.api as sm
```

```python
X_train_lm = sm.add_constant(X_train_rfe)
lm = sm.OLS(y_train,X_train_lm).fit()
print(lm.summary())
```

```python
lm
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
X = X_train_lm
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import r2_score
from sklearn.metrics import make_scorer
```

```python
model = DecisionTreeRegressor(random_state=100)
params = {'max_depth':[2,3,4,5,10,20,25,30],
          'min_samples_leaf':[1,5,10,20,30,50,70,100],
          'max_features': [0.35,0.4,0.45,0.5]
          }
grid_search = GridSearchCV(estimator=model,param_grid=params,n_jobs=-1,verbose=1,scoring='r2')
grid_search.fit(X_train,y_train)
```

```python
model_best = grid_search.best_estimator_
```

```python
y_train_pred = model_best.predict(X_train)
y_train_pred
```

```python
print(r2_score(y_train,y_train_pred))
```

```python
y_test_pred = model_best.predict(X_test)
y_test_pred
```

```python
print(r2_score(y_test,y_test_pred))
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor(random_state=100)
params = {'max_depth':[2,3,4,5,10,20,30],
          'min_samples_leaf':[1,5,10,20],
          'n_estimators':[40,50,100],
          'max_features': [0.3,0.4,0.5,0.6]}

grid_search = GridSearchCV(estimator=model,param_grid=params,n_jobs=-1,verbose=1,scoring='r2')
grid_search.fit(X_train,y_train)
```

```python
model_best = grid_search.best_estimator_
```

```python
y_train_pred = model_best.predict(X_train)
y_train_pred
```

```python
print(r2_score(y_train,y_train_pred))
```

```python
y_test_pred = model_best.predict(X_test)
y_test_pred
```

```python
print(r2_score(y_test,y_test_pred))
```

```python
model_best.feature_importances_
```
